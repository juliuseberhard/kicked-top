```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# number of iterations
N = 1000

# number of initial values per component
R = 11

# parameters
alpha = 1.2
tau = 2

# components of angular momentum
# initial conditions
jx, jz = np.meshgrid(np.linspace(-1, 1, R), np.linspace(-1, 1, R))
jy = np.where(jx**2 + jz**2 <= 1, np.sqrt(abs(1 - jx**2 - jz**2)), np.nan)

# initialize plot
fig = plt.figure()
plt.plot(jx[~np.isnan(jy)], jz[~np.isnan(jy)], '.', color='k', ms=0.5)

# loop
for i in range(1, N):
    X = np.array(jx)
    A = jy*np.cos(alpha)
    B = jz*np.sin(alpha)
    C = jy*np.sin(alpha)
    D = jz*np.cos(alpha)
    Y = A-B
    Z = C+D
    jx = X*np.cos(tau*Z) - Y*np.sin(tau*Z)
    jy = X*np.sin(tau*Z) + Y*np.cos(tau*Z)
    jz = Z
    jx_plot = np.where(jy[~np.isnan(jy)]>=0, jx[~np.isnan(jy)], np.nan)
    jz_plot = np.where(jy[~np.isnan(jy)]>=0, jz[~np.isnan(jy)], np.nan)
    plt.plot(jx_plot, jz_plot, '.', color='k', ms=0.5)

plt.axis('square')
#plt.xlim([-1, 1])
#plt.ylim([-1, 1])
```
