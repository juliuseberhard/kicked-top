### 2 Grundlagen

**Kreisel.**
Ein Kreisel ist ein starrer Körper, der sich dreht.
Im Gravitationsfeld wird häufig ein *kräftefreier* Kreisel, auf den keinerlei resultierende Drehmomente einwirken, von einem *schweren* Kreisel unterschieden, dessen Bewegung durch dauerhafte äußere Krafteinwirkung bestimmt ist [2].
Der hier diskutierte gekickte Kreisel zeichnet sich durch eine kräftefreie Bewegung aus, die durch periodische Stöße gestört wird.
Jeder Kreisel besitzt eine körperfeste Achse, mit der ein maximales Trägheitsmoment assoziiert ist, und senkrecht dazu eine Achse mit dem minimalen Trägheitsmoment.
Diese zwei Achsen und die zu beiden senkrecht stehende körperfeste Achse definieren die Hauptträgheitsmomente eines Kreisels.
Beim hier betrachteten *symmetrischen* Kreisel sind die beiden größeren Hauptträgheitsmomente gleich.

**Nutation.**
Die wesentliche Invariante beim kräftefreien Kreisel ist der vektorielle Drehimpuls $\underline{L}$:
$$
\dot{\underline{L}} = 0.
$$
Nachdem ein kräftefreier symmetrischer Kreisel in Rotation um seine Symmetrieachse versetzt wird, liegt $\underline{L}$ genau auf dieser Achse.
Ein zusätzlicher, kurzer Stoß außerhalb des Schwerpunkts führt dazu, dass die Lage der Rotationsachse nicht mehr mit der Richtung des Drehimpulses übereinstimmt.
Die Drehimpulserhaltung macht sich dadurch bemerkbar, dass die Achse des Kreisels um den beim Stoß leicht veränderten, aber danach wieder konstanten Drehimpulsvektor $\underline{L}'$ herum *nutiert*, das heißt, die Endpunkte der Achse bewegen sich in konzentrischen Kreisen um $\underline{L}'$.
Eine solche Nutation ist die Ausgangssituation für den gekickten Kreisel.

**Koordinatensystem.**
Die Beschreibung der Kreiselbewegung erfolgt mithilfe eines kartesischen Koordinatensystems, welches sich an der Position des Kreisels im Raum orientiert:
Die x-Achse zeigt immer entlang der Symmetrieachse des Kreisels und behält ihre Richtung bezogen auf den Drehsinn des Kreisels bei; dreht sich die Symmetrieachse des rotierenden Kreisels um 180°, ändert sich auch die x-Richtung.
Die z-Achse liegt in der Ebene, die durch die körperfeste x-Achse und die raumfeste Vertikale aufgespannt wird.
Die y-Achse komplettiert das x-y-z-Rechtssystem.
Die Orientierung von y- und z-Richtung ist beliebig, bleibt aber immer konsistent mit der Orientierung der x-Richtung.
Eine Nutation des Kreisels um $\underline{L}$ wird im kreiseleigenen Koordinatensystem als „Nutation“ von $\underline{L}$ um die x-Achse beschrieben.

**Kicks.**
Zeitlich periodisch wird der Kreisel so außerhalb seines Schwerpunkts gestoßen (gekickt), dass er ein Drehmoment $\underline{M}$ erhält und eine Drehung um seine momentane z-Achse ausführt.
Unter der Annahme, dass ein Kick keine zeitliche Ausdehnung besitzt, ändert sich gemäß
$$
d \underline{L} = \underline{M} d t
$$
der Drehimpuls des Kreisels nicht, also bleibt dessen Betrag während des gesamten Prozesses konstant,
$$
\underline{L}^2 = const,
$$
und die Spitze von $\underline{L}$ liegt auf einer Kugeloberfläche.

### 3 Modell

**Nutation und Kicks.**
Es wird Folgendes festgelegt [1]:
- Die Nutation des Kreisels um $\underline{L}$ erfolgt mit konstanter Winkelgeschwindigkeit; das entspricht der normalen kräftefreien Nutation.
- Ein Kick bewirkt eine instantane Drehung des Kreisels um die momentane z-Achse.
- Der durch einen Kick hervorgerufene Drehwinkel ist proportional zur z-Komponente des Drehimpulses $L_z$.

**Stroboskopische Abbildung.**
Die Bewegung des Drehimpulsvektors im kreiseleigenen Koordinatensystem wird durch eine stroboskopische Abbildung
$$
\underline{L}(t) \mapsto \underline{L}(t + T)
$$
beschrieben.
$T$ entspricht der Periodendauer zwischen zwei Kicks.
Die Abbildung soll zu einem Zeitpunkt starten, der sich unmittelbar nach einem Kick befindet.
Jedes Bildelement der Abbildung ist demnach Ergebnis der Nutation des Kreisels seit dem zuletzt registrierten Zustand und des darauffolgenden Kicks.

**Parameter.** 
Zur Beschreibung der Kreiselbewegung werden zwei Parameter herangezogen: die Winkelgeschwindigkeit der Nutation um die x-Achse $\alpha$ und die *Kickstärke* $\tau$.
Die Kickstärke ist der Proportionalitätsfaktor zwischen $L_z$ und dem Drehwinkel eines Kicks und repräsentiert das durch den Kick erzeugte Drehmoment am Kreisel.

**Konstanten.**
Zur Vereinfachung der Abbildungsvorschrift wird
$$
\underline{L}^2 = 1
$$
und
$$
T = 1
$$
gesetzt.
Damit liegt die Bildmenge der Abbildung auf einer Einheitssphäre, $\alpha$ entspricht dem Nutationswinkel pro Periode und $\tau L_z$ dem Drehwinkel pro Kick.

**Abbildungsvorschrift.**
Die stroboskopische Abbildung wird in zwei Abbildungen aufgeteilt:
$$
\underline{L}(t) \mapsto \underline{\tilde{L}} \mapsto \underline{L}(t + T).
$$
Die erste entspricht der Nutation um den Winkel $\alpha$, die zweite der Drehung um $\tau L_z$ nach dem Kick.
Da beide Bewegungen als nacheinander und isoliert ablaufend betrachtet werden können, reicht es aus, die beiden geometrisch hergeleiteten Abbildungsvorschriften zu verknüpfen.

$\underline{L}(t) \mapsto \underline{\tilde{L}}$:
$$
\tilde{L}_x = L_x(t)
$$
$$
\tilde{L}_y = L_y(t) \cos \alpha - L_z(t) \sin \alpha
$$
$$
\tilde{L}_z = L_z(t) \cos \alpha + L_y(t) \sin \alpha
$$

$\underline{\tilde{L}} \mapsto \underline{L}(t + T)$:
$$
L_x(t + T) = \tilde{L}_x \cos \lambda \tilde{L}_z - \tilde{L}_y \sin \lambda \tilde{L}_z
$$
$$
L_y(t + T) = \tilde{L}_y \cos \lambda \tilde{L}_z + \tilde{L}_x \sin \lambda \tilde{L}_z
$$
$$
L_z(t + T) = \tilde{L}_z
$$

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# number of iterations
N = 1000

# number of initial values per component
R = 7

# parameters
alpha = 2.7
tau = 3

# components of angular momentum
# initial conditions
Lx, Lz = np.meshgrid(np.linspace(-1, 1, R), np.linspace(-1, 1, R))
Ly = np.where(Lx**2 + Lz**2 <= 1, np.sqrt(abs(1 - Lx**2 - Lz**2)), np.nan)

# initialize plot
fig = plt.figure()
plt.plot(Lx[~np.isnan(Ly)], Lz[~np.isnan(Ly)], '.', color='r', ms=10)

# loop
for i in range(1, N):
    X = np.array(Lx)
    A = Ly*np.cos(alpha)
    B = Lz*np.sin(alpha)
    C = Ly*np.sin(alpha)
    D = Lz*np.cos(alpha)
    Y = A-B
    Z = C+D
    Lx = X*np.cos(tau*Z) - Y*np.sin(tau*Z)
    Ly = X*np.sin(tau*Z) + Y*np.cos(tau*Z)
    Lz = Z
    Lx_plot = np.where(Ly[~np.isnan(Ly)]>=0, Lx[~np.isnan(Ly)], np.nan)
    Lz_plot = np.where(Ly[~np.isnan(Ly)]>=0, Lz[~np.isnan(Ly)], np.nan)
    plt.plot(Lx_plot, Lz_plot, '.', color='k', ms=0.5)
    plt.xlabel('L_x')
    plt.ylabel('L_z')

plt.axis('square')
#plt.xlim([-1, 1])
#plt.ylim([-1, 1])
```
