### 2 Grundlagen

**Kreisel.**
Ein Kreisel ist ein starrer Körper, der sich dreht.
Im Gravitationsfeld wird häufig ein *kräftefreier* Kreisel, auf den keinerlei resultierende Drehmomente einwirken, von einem *schweren* Kreisel unterschieden, dessen Bewegung durch dauerhafte äußere Krafteinwirkung bestimmt ist [2].
Der hier diskutierte gekickte Kreisel zeichnet sich durch eine kräftefreie Bewegung aus, die durch periodische Stöße gestört wird.
Jeder Kreisel besitzt eine körperfeste Achse, mit der ein maximales Trägheitsmoment assoziiert ist, und senkrecht dazu eine Achse mit dem minimalen Trägheitsmoment.
Diese zwei Achsen und die zu beiden senkrecht stehende körperfeste Achse definieren die Hauptträgheitsmomente eines Kreisels.
Beim hier betrachteten *symmetrischen* Kreisel sind die beiden größeren Hauptträgheitsmomente gleich [2].

**Nutation.**
Die wesentliche Invariante beim kräftefreien Kreisel ist der vektorielle Drehimpuls $\underline{L}$:
$$
\dot{\underline{L}} = 0.
$$
Nachdem ein kräftefreier symmetrischer Kreisel in Rotation um seine Symmetrieachse versetzt wird, liegt $\underline{L}$ genau auf dieser Achse.
Ein zusätzlicher, kurzer Stoß außerhalb des Schwerpunkts führt dazu, dass die Lage der Rotationsachse nicht mehr mit der Richtung des Drehimpulses übereinstimmt.
Die Drehimpulserhaltung macht sich dadurch bemerkbar, dass die Achse des Kreisels um den beim Stoß leicht veränderten, aber danach wieder konstanten Drehimpulsvektor $\underline{L}'$ herum *nutiert*, das heißt, die Endpunkte der Achse bewegen sich in konzentrischen Kreisen um $\underline{L}'$ [2].
Eine solche Nutation ist die Ausgangssituation für den gekickten Kreisel.

**Koordinatensystem.**
Die Beschreibung der Kreiselbewegung erfolgt mithilfe eines kartesischen Koordinatensystems, welches sich an der Position des Kreisels im Raum orientiert:
Die x-Achse zeigt immer entlang der Symmetrieachse des Kreisels und behält ihre Richtung bezogen auf den Drehsinn des Kreisels bei; dreht sich die Symmetrieachse des rotierenden Kreisels um 180°, ändert sich auch die x-Richtung.
Die z-Achse liegt in der Ebene, die durch die körperfeste x-Achse und die raumfeste Vertikale aufgespannt wird.
Die y-Achse komplettiert das x-y-z-Rechtssystem.
Die Orientierung von y- und z-Richtung ist beliebig, bleibt aber immer konsistent mit der Orientierung der x-Richtung.
Eine Nutation des Kreisels um $\underline{L}$ wird im kreiseleigenen Koordinatensystem folglich als „Nutation“ von $\underline{L}$ um die x-Achse beschrieben.

**Kicks.**
Zeitlich periodisch wird der Kreisel so außerhalb seines Schwerpunkts gestoßen (gekickt), dass er ein Drehmoment $\underline{M}$ erhält und eine Drehung um seine momentane z-Achse ausführt.
Unter der Annahme, dass ein Kick keine zeitliche Ausdehnung besitzt, ändert sich gemäß
$$
d \underline{L} = \underline{M} d t
$$
der Drehimpuls des Kreisels nicht, also bleibt dessen Betrag während des gesamten Prozesses konstant,
$$
\underline{L}^2 = const,
$$
und die Spitze von $\underline{L}$ liegt auf einer Kugeloberfläche [1].

**Reguläres und chaotisches Verhalten.**
Die Bewegung des gekickten Kreisels folgt deterministischen Regeln, sodass die Bewegung des Kreisels aus einer festen Anfangsbedingung heraus immer dieselbe Trajektorie auf der Einheitskugel zeigt.
Führt eine infinitesimale Änderung dieser Anfangsbedingung zu einer endlichen Änderung der Trajektorie, spricht man von chaotischem Verhalten, andernfalls von regulärem Verhalten.
Numerisch lässt sich das Verhalten untersuchen, indem die Bewegung für mehrere sehr dicht beieinanderliegende Anfangsbedingungen berechnet wird und die Trajektorien miteinander veglichen werden.

### 3 Modell

**Nutation und Kicks.**
Es wird Folgendes festgelegt [1]:
- Die Nutation des Kreisels um $\underline{L}$ erfolgt mit konstanter Winkelgeschwindigkeit; das entspricht der normalen kräftefreien Nutation.
- Ein Kick bewirkt eine instantane Drehung des Kreisels um die momentane z-Achse.
- Der durch einen Kick hervorgerufene Drehwinkel ist proportional zur z-Komponente des Drehimpulses $L_z$.

**Stroboskopische Abbildung.**
Die Bewegung des Drehimpulsvektors im kreiseleigenen Koordinatensystem wird durch eine stroboskopische Abbildung
$$
\underline{L}(t) \mapsto \underline{L}(t + T)
$$
beschrieben.
$T$ entspricht der Periodendauer zwischen zwei Kicks.
Die Abbildung soll zu einem Zeitpunkt starten, der sich unmittelbar nach einem Kick befindet.
Jedes Bildelement der Abbildung ist demnach Ergebnis der Nutation des Kreisels seit dem zuletzt registrierten Zustand und des darauffolgenden Kicks.

**Parameter.** 
Zur Beschreibung der Kreiselbewegung werden zwei Parameter benötigt: die Winkelgeschwindigkeit der Nutation um die x-Achse $\alpha$ und die *Kickstärke* $\tau$.
Die Kickstärke ist der Proportionalitätsfaktor zwischen $L_z$ und dem Drehwinkel eines Kicks und repräsentiert das durch den Kick erzeugte Drehmoment am Kreisel.

**Konstanten.**
Zur Vereinfachung der Abbildungsvorschrift wird
$$
\underline{L}^2 = 1
$$
und
$$
T = 1
$$
gesetzt.
Damit liegt die Bildmenge der Abbildung auf einer Einheitssphäre, $\alpha$ entspricht dem Nutationswinkel pro Periode und $\tau L_z$ dem Drehwinkel pro Kick.

**Abbildungsvorschrift.**
Die stroboskopische Abbildung wird in zwei Abbildungen aufgeteilt:
$$
\underline{L}(t) \mapsto \underline{\tilde{L}} \mapsto \underline{L}(t + T).
$$
Die erste entspricht der Nutation um den Winkel $\alpha$, die zweite der Drehung um $\tau L_z$ durch den Kick.
Da beide Bewegungen als nacheinander und isoliert ablaufend betrachtet werden können, reicht es aus, die beiden geometrisch hergeleiteten Abbildungsvorschriften zu verknüpfen.

$\underline{L}(t) \mapsto \underline{\tilde{L}}$:
$$
\tilde{L}_x = L_x(t)
$$
$$
\tilde{L}_y = L_y(t) \cos \alpha - L_z(t) \sin \alpha
$$
$$
\tilde{L}_z = L_z(t) \cos \alpha + L_y(t) \sin \alpha
$$

$\underline{\tilde{L}} \mapsto \underline{L}(t + T)$:
$$
L_x(t + T) = \tilde{L}_x \cos \tau \tilde{L}_z - \tilde{L}_y \sin \tau \tilde{L}_z
$$
$$
L_y(t + T) = \tilde{L}_y \cos \tau \tilde{L}_z + \tilde{L}_x \sin \tau \tilde{L}_z
$$
$$
L_z(t + T) = \tilde{L}_z
$$


### 4 Python-Code

**Funktionen.**
Es werden 3 Funktionen definiert:
- `sign_no_zero` berechnet die Signumfunktion des reellen Arguments, wobei $\text{sgn}(x) = 1$, wenn $x = 0$, und wird in der Funktion `kicked_top_ad` verwendet.
- `kicked_top_map` plottet eine Hemisphäre ($L_y \geq$ 0) der stroboskopischen Abbildung der Bewegung des gekickten Kreisels für gegebene Anfangsbedingungen (= anfängliche Positionen des Drehimpulsvektors), Anfangsbedingungen werden rot markiert.
- `kicked_top_ad` berechnet die pro Periode zurückgelegte Winkeldistanz des Drehimpulses auf der Einheitskugel für gegebene Anfangsbedingungen.

```python
def sign_no_zero(x):
    return(np.where(np.sign(x)==0, 1, np.sign(x)))

#

def kicked_top_map(angv, kick, iter_num, init_cond, ax, ms):
    '''
    Bildet periodisch die Position des Drehimpulsvektors
    eines gekickten Kreisels auf der Einheitskugel ab.
    Periode = 1
    # Parameter:
    angv:      Nutationswinkel um x-Achse pro Periode
    kick:      Kickstaerke
    iter_num:  Zahl der abzubildenden Perioden
    init_cond: Liste von 2 Vektoren mit x- & z-Komponenten
               der Anfangsbedingungen fuer Position des Drehimpulses
    ax:        Achse fuer Plot
    ms:        marker size
    # Ausgabe:
    x-z-Plot der stroboskopischen Abbildung L(t) -> L(t+1)
    '''

    # Anfangsbedingungen Drehimpuls
    Lx, Lz = np.meshgrid(init_cond[0], init_cond[1])
    # (L liegt auf Einheitskugel)
    Ly = np.where(Lx**2+Lz**2 <= 1, np.sqrt(abs(1 - Lx**2 - Lz**2)), np.nan)

    # Plot initialisieren, Anfangsbedingungen markieren
    ax.plot(Lx[~np.isnan(Ly)], Lz[~np.isnan(Ly)], '.', color='r', ms=5)

    # Iterationen
    for i in range(1, iter_num):
        # Abbildungsvorschrift
        X = np.array(Lx)
        A = Ly*np.cos(angv)
        B = Lz*np.sin(angv)
        C = Ly*np.sin(angv)
        D = Lz*np.cos(angv)
        Y = A-B
        Z = C+D
        Lx = X*np.cos(kick*Z) - Y*np.sin(kick*Z)
        Ly = X*np.sin(kick*Z) + Y*np.cos(kick*Z)
        Lz = Z
        Lx_plot = np.where(Ly[~np.isnan(Ly)]>=0, Lx[~np.isnan(Ly)], np.nan)  # nur innerhalb der Kugel plotten
        Lz_plot = np.where(Ly[~np.isnan(Ly)]>=0, Lz[~np.isnan(Ly)], np.nan)  # dito
        ax.plot(Lx_plot, Lz_plot, '.', color='k', ms=ms)
        ax.set_xlabel('')
        ax.set_ylabel('')

    # quadratische Plots
    ax.set(adjustable='box-forced', aspect='equal')
    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])

#

def kicked_top_ad(angv, kick, iter_num, init_cond):
    '''
    Berechnet die Winkeldistanz zwischen zwei Kicks
    beim gekickten Kreisel.
    Berechnung nach en.wikipedia.org/wiki/Angular_distance
    # Parameter:
    angv:      Nutationswinkel um x-Achse pro Periode
    kick:      Kickstaerke
    iter_num:  Zahl der abzubildenden Perioden
    init_cond: Liste von 2 Vektoren mit x- & z-Komponenten
               der Anfangsbedingungen fuer Position des Drehimpulses
    # Ausgabe:
    Array (len(init_cond[0], len(init_cond[1]), iter_num) der Winkeldistanzen
    '''
    # Anfangsbedingungen Drehimpuls
    Lx, Lz = np.meshgrid(init_cond[0], init_cond[1])
    Ly = np.where(Lx**2+Lz**2 <= 1, np.sqrt(abs(1 - Lx**2 - Lz**2)), np.nan)

    # Winkeldistanz-Array
    ang_dist = np.empty((init_cond[0].shape[0], init_cond[1].shape[0], iter_num))

    # Iterationen
    for i in range(1, iter_num):
        # Rektaszension und Deklination vor Kick
        r1 = (1 - sign_no_zero(Lx))*np.pi/2 + sign_no_zero(Lx)*np.arccos(Ly)
        d1 = np.arcsin(Lz)
        # Abbildungsvorschrift
        X = np.array(Lx)
        A = Ly*np.cos(angv)
        B = Lz*np.sin(angv)
        C = Ly*np.sin(angv)
        D = Lz*np.cos(angv)
        Y = A-B
        Z = C+D
        Lx = X*np.cos(kick*Z) - Y*np.sin(kick*Z)
        Ly = X*np.sin(kick*Z) + Y*np.cos(kick*Z)
        Lz = Z
        # Rektaszension und Deklination nach Kick
        r2 = (1 - sign_no_zero(Lx))*np.pi/2 + sign_no_zero(Lx)*np.arccos(Ly)
        d2 = np.arcsin(Lz)
        # Winkeldistanz
        ad = np.arccos(np.sin(d1)*np.sin(d2) + np.cos(d1)*np.cos(d2)*np.cos(r1-r2))
        ang_dist[:, :, i] = ad

    return(ang_dist)
```

**Übergang von regulärer zu chaotischer Bewegung.**
Mit der Funktion `kicked_top_map` wird die Kreiselbewegung für 49 Anfangsbedingungen ($L_x$ und $L_z$ zwischen 0 und 1) und 36 Parameter-Kombinationen ($\alpha$ zwischen 0 und 7.5, $\tau$ zwischen 0 und 5) geplottet.
Für $\alpha = 0$ und/oder $\tau = 0$ ist zu beobachten, dass eine Nutation bzw. Drehung des Kreisels nur um die Achse erfolgt, deren assoziierter Parameter ($\alpha$ zur x-, $\tau$ zur z-Achse) nicht null ist.
Eine Erhöhung der Nutationsgeschwindigkeit bei konstanter Kickstärke führt nur bei großen Kickstärken zu einer leichten Ausweitung der chaotischen Regime.
Jedoch kann die Nutation zwischen zwei Kicks dazu führen, dass der Kreisel trotz hoher Kickstärke deutlich erkennbare reguläre Regimes aufweist.
Dies ist bei Werten nahe $\pi$ ($\alpha = 3, 6$) zu beobachten.
Zunehmende Kickstärken führen einigermaßen zuverlässig zu chaotischerem Verhalten.

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# Figure-Groesse im Notebook
plt.rcParams['figure.figsize'] = (12., 12.)

# Parameter
alpha = np.linspace(0, 7.5, 6)  # Winkelgeschw. Nutation
tau = np.linspace(0, 5, 6)  # Kickstärke

# Anzahl Iterationen
N = 500

# Anzahl Anfangsbedingungen pro L-Komponente
R = 7

# Anfangsbedingungen x, z
Lx_vec = np.linspace(-1, 1, R)
Lz_vec = np.linspace(-1, 1, R)

# Abbildung plotten
f, ax = plt.subplots(tau.shape[0], alpha.shape[0], sharex=True, sharey=True)
for i in range(alpha.shape[0]):
    for j in range(tau.shape[0]):
        kicked_top_map(alpha[i], tau[j], N, [Lx_vec, Lz_vec], ax=ax[j, i], ms=0.5)

f.subplots_adjust(hspace=0.2, wspace=0.2)

f.text(0.07, 0.6, r'$L_z$', fontsize=16)
f.text(0.5, 0.07, r'$L_x$', fontsize=16)

f.text(0.16, 0.92, r'$\alpha = $%.1f' %alpha[0])
f.text(0.295, 0.92, r'$\alpha = $%.1f' %alpha[1])
f.text(0.43, 0.92, r'$\alpha = $%.1f' %alpha[2])
f.text(0.565, 0.92, r'$\alpha = $%.1f' %alpha[3])
f.text(0.7, 0.92, r'$\alpha = $%.1f' %alpha[4])
f.text(0.835, 0.92, r'$\alpha = $%.1f' %alpha[5])

f.text(0.92, 0.81, r'$\tau = $%.1f' %tau[0])
f.text(0.92, 0.68, r'$\tau = $%.1f' %tau[1])
f.text(0.92, 0.55, r'$\tau = $%.1f' %tau[2])
f.text(0.92, 0.42, r'$\tau = $%.1f' %tau[3])
f.text(0.92, 0.29, r'$\tau = $%.1f' %tau[4])
f.text(0.92, 0.16, r'$\tau = $%.1f' %tau[5])

plt.show()
```

**Feine Variation der Anfangsbedingungen.**
Da der Übergang von regulärem zu chaotischem Verhalten des gekickten Kreisels vor allem mit einer Erhöhung der Kickstärke einher zu gehen scheint, soll dieser Übergang durch eine leichte Veränderung der Kickstärke genauer visualisiert werden.
Bei $\alpha$ = 4.5 wird $\tau$ von 2.3 über 2.35 auf 2.4 erhöht.
Um „reguläres“ und „chaotisches“ Verhalten zu unterscheiden, betrachte ich 100 sehr nah beieinander liegende Anfangsbedingungen (10 Werte von $L_x$ zwischen 0 und 0.001, 10 von $L_z$ zwischen 0.5 und 0.501), in den Plots rot markiert.
Bei $\tau$ = 2.3 zeigt die zeitliche Entwicklung (über 10, 100 und 500 Perioden) einen festen Orbit, den $\underline{L}$ nicht verlässt.
Bei $\tau$ = 2.35 ist immer noch ein Orbit zu erkennen, neben dem aber nach längerer Zeit ein zweiter, kleinerer erscheint.
Nach 500 Perioden haben sich viele Punkte deutlich sichtbar vom ursprünglichen Orbit entfernt.
Bei $\tau$ = 2.4 lässt sich nach 10 Perioden zwar ein ähnlicher Orbit erahnen, der aber schon nach 100 Perioden nicht mehr eindeutig lokalisierbar ist.
Die starke Streuung der anfangs dicht beieinander liegenden Punkte lässt auf einen Übergang zum chaotischen Verhalten ab $\tau \approx$ 2.35 schließen.

```python
# Parameter
alpha = 4.5
tau = np.linspace(2.3, 2.4, 3)

# Anzahl Iterationen
N = np.array([10, 100, 500])

# Anzahl Anfangsbedingungen pro L-Komponente
R = 10

# Anfangsbedingungen x, z
Lx_vec = np.linspace(0, 0.001, R)
Lz_vec = np.linspace(0.5, 0.501, R)

# Abbildung plotten
f, ax = plt.subplots(N.shape[0], tau.shape[0], sharex=True, sharey=True)
for i in range(tau.shape[0]):
    for j in range(N.shape[0]):
        kicked_top_map(alpha, tau[i], N[j], [Lx_vec, Lz_vec], ax=ax[j, i], ms=0.5)

f.text(0.07, 0.6, r'$L_z$', fontsize=16)
f.text(0.5, 0.07, r'$L_x$', fontsize=16)

f.text(0.22, 0.92, r'$\tau = $%.2f' %tau[0])
f.text(0.495, 0.92, r'$\tau = $%.2f' %tau[1])
f.text(0.77, 0.92, r'$\tau = $%.2f' %tau[2])

f.text(0.92, 0.785, r'$N = $%d' %N[0])
f.text(0.92, 0.51, r'$N = $%d' %N[1])
f.text(0.92, 0.235, r'$N = $%d' %N[2])

plt.show()
```

Auch nach 1000 Perioden ist die Präferenz dieser Orbitale sehr deutlich.
Für $\tau$ = 2.7 streuen die Punkte von Beginn an stärker als für die niedrigeren Kickstärken und nach 1000 Perioden nimmt die Bewegung von $\underline{L}$ fast die gesamte Kugeloberfläche ein, das heißt, die Bewegung ist sehr chaotisch.

```python
# Parameter
alpha = 4.5
tau = np.linspace(2.3, 2.7, 3)  # 2.25, 2.4

# Anzahl Iterationen
N = np.array([100, 500, 1000])

# Anzahl Anfangsbedingungen pro L-Komponente
R = 10

# Anfangsbedingungen x, z
Lx_vec = np.linspace(0.5, 0.501, R)
Lz_vec = np.linspace(0.5, 0.501, R)

# Abbildung plotten
f, ax = plt.subplots(N.shape[0], tau.shape[0], sharex=True, sharey=True)
for i in range(tau.shape[0]):
    for j in range(N.shape[0]):
        kicked_top_map(alpha, tau[i], N[j], [Lx_vec, Lz_vec], ax[j, i], ms=0.5)

f.text(0.07, 0.6, r'$L_z$', fontsize=16)
f.text(0.5, 0.07, r'$L_x$', fontsize=16)

f.text(0.22, 0.92, r'$\tau = $%.1f' %tau[0])
f.text(0.495, 0.92, r'$\tau = $%.1f' %tau[1])
f.text(0.77, 0.92, r'$\tau = $%.1f' %tau[2])

f.text(0.92, 0.785, r'$N = $%d' %N[0])
f.text(0.92, 0.51, r'$N = $%d' %N[1])
f.text(0.92, 0.235, r'$N = $%d' %N[2])

plt.show()
```

**Visualisierung von regulärem und chaotischem Verhalten mittels Winkeldistanz.**
Reguläres und chaotisches Verhalten zu quantifizieren, ist beim gekickten Kreisel durch die Betrachtung der zwischen zwei Kicks zurückgelegten Winkeldistanz des Drehimpulsvektors möglich:
Mit regulärer Kreiselbewegung gehen Orbitale einher, die $\underline{L}$ nicht verlässt.
Die Änderung des Winkels auf der Einheitskugel wiederholt sich periodisch und innerhalb eines kleinen Spielraums, da immer wieder die gleichen Ausgangsbedingungen eingenommen werden.
Für chaotisches Verhalten ist die Winkeldistanz zwischen zwei Kicks nicht periodisch, da die Bewegung nicht auf Orbitalen erfolgt.
Im folgenden Code werden für 4 Anfangsbedingungen bei $\alpha$ = 4.5, $\tau$ = 2.5 die Bewegungen von $\underline{L}$ geplottet (Funktion `kicked_top_map`) und die zurückgelegten Winkeldistanzen pro Kick berechnet (`kicked_top_ad`) und für 1000 Perioden geplottet.
Anfangsbedingung 1 führt zu einem stabilen Orbit und einer periodischen, eng begrenzten Variation der Winkeldistanz.
Anfangsbedingung 4 hat weniger stabile Orbits zur Folge und zeigt ab 750 Perioden eine beginnende chaotische Bewegung.
Die chaotischen Bewegungen infolge der Anfangsbedingungen 2 und 3 spiegeln sich in den Plots der Winkeldistanzen wider, wobei auch dort kurzzeitige periodische Orbitalbewegungen erkennbar sind (z. B. Anfangsbed. 2 zwischen 300 und 450 Perioden).
Durch eine Fourieranalyse der Winkeldistanz-Zeitreihen ließe sich vielleicht ein Maß für die Regularität der Bewegung finden.

```python
# Parameter
alpha = 4.5
tau = 2.5

# Anzahl Iterationen
N = 1000

# Anfangsbedingungen x, z
Lx_vec = np.array([-0.5, 0.5])
Lz_vec = np.array([-0.5, 0.5])

# Abbildung plotten
f1, ax1 = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(Lx_vec.shape[0]):
    for j in range(Lz_vec.shape[0]):
        kicked_top_map(alpha, tau, N, [Lx_vec[i], Lz_vec[j]], ax1[i, j], ms=1)

f1.subplots_adjust(hspace=0.1, wspace=0.1)

f1.suptitle(r'$\alpha = 4.5$, $\tau = 2.5$', y=0.94, fontsize=16)

f1.text(0.5, 0.08, r'$L_x$', fontsize=16)
f1.text(0.05, 0.5, r'$L_z$', fontsize=16)

ax1[0, 0].text(0, 0, 'Anfangsbedingung 1', ha='center')
ax1[0, 1].text(0, 0, 'Anfangsbedingung 2', ha='center')
ax1[1, 0].text(0, 0, 'Anfangsbedingung 3', ha='center')
ax1[1, 1].text(0, 0, 'Anfangsbedingung 4', ha='center')

# Winkeldistanzen berechnen
ad_array = kicked_top_ad(alpha, tau, N, [Lx_vec, Lz_vec])

# Winkeldistanzen plotten
f2, ax2 = plt.subplots(4, 1, sharex=True)
ax2[0].plot(range(N), ad_array[0, 0, :])
ax2[0].set_ylim([0, 2.5])
ax2[1].plot(range(N), ad_array[0, 1, :])
ax2[1].set_ylim([0, 2.5])
ax2[2].plot(range(N), ad_array[1, 0, :])
ax2[2].set_ylim([0, 2.5])
ax2[3].plot(range(N), ad_array[1, 1, :])
ax2[3].set_ylim([0, 2.5])

ax2[0].text(N/2, 2.2, 'Anfangsbedingung 1', ha='center')
ax2[1].text(N/2, 2.2, 'Anfangsbedingung 2', ha='center')
ax2[2].text(N/2, 2.2, 'Anfangsbedingung 3', ha='center')
ax2[3].text(N/2, 2.2, 'Anfangsbedingung 4', ha='center')

f2.text(0.5, 0.05, 'Periode')
f2.text(0.05, 0.5, 'Winkeldistanz zwischen zwei Kicks', rotation=90)

plt.show()
```
